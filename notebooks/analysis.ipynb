```python

# count files for each dataset
import os
import pandas as pd

def count_files_in_subdirectories(base_path):
    num_files = {
        'vt': {},
        'oss-detect-backdoor': {},
        'bincapz': {},
        'packj': {}
    }
    
    # Iterate through the first level of directories
    for level1 in os.listdir(base_path):
        level1_path = os.path.join(base_path, level1)
        
        if os.path.isdir(level1_path) and level1 in num_files:
            # Iterate through the second level of directories
            for level2 in os.listdir(level1_path):
                level2_path = os.path.join(level1_path, level2)
                
                if os.path.isdir(level2_path):
                    # Check if there are third level directories
                    third_level_dirs = [d for d in os.listdir(level2_path) if os.path.isdir(os.path.join(level2_path, d))]
                    
                    if third_level_dirs:
                        # Iterate through the third level of directories
                        for level3 in third_level_dirs:
                            level3_path = os.path.join(level2_path, level3)
                            
                            if os.path.isdir(level3_path):
                                # Count the number of files in the third level directory
                                if level1 == 'packj':
                                    # Only count .json files
                                    json_files = [f for f in os.listdir(level3_path) if f.endswith('.json') and os.path.isfile(os.path.join(level3_path, f))]
                                    num_files[level1][f"{level2}-{level3}"] = len(json_files)
                                else:
                                    # Count all files
                                    all_files = [f for f in os.listdir(level3_path) if os.path.isfile(os.path.join(level3_path, f))]
                                    num_files[level1][f"{level2}-{level3}"] = len(all_files)
                    else:
                        # Count the number of files in the second level directory
                        if level1 == 'packj':
                            # Only count .json files
                            json_files = [f for f in os.listdir(level2_path) if f.endswith('.json') and os.path.isfile(os.path.join(level2_path, f))]
                            num_files[level1][level2] = len(json_files)
                        else:
                            # Count all files
                            all_files = [f for f in os.listdir(level2_path) if os.path.isfile(os.path.join(level2_path, f))]
                            num_files[level1][level2] = len(all_files)
    
    return num_files


base_path = r'../scan-results'
file_counts = count_files_in_subdirectories(base_path)


df = pd.DataFrame(file_counts).fillna(0).astype(int)

# Reset the index to ensure the first row contains {level2}-{level3} or {level2}
df.reset_index(inplace=True)
df.rename(columns={'index': 'Dataset'}, inplace=True)

# Display the DataFrame
print(df)


```
